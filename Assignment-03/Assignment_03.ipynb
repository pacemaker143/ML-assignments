```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/content/Admission_Predict_Ver1.1.csv')
data.describe()
```

```python
# 1. Scatter Plot: GRE Score vs. CGPA
plt.figure(figsize=(8, 6))
sns.scatterplot(x='GRE Score', y='CGPA', hue='Chance of Admit ', data=data)
plt.title('GRE Score vs. CGPA')
plt.show()
```

```python
# 2. Bar Plot: University Rating vs. Average Chance of Admit
university_rating_avg_chance = data.groupby('University Rating')['Chance of Admit '].mean()
plt.figure(figsize=(8, 6))
university_rating_avg_chance.plot(kind='bar')
plt.title('Average Chance of Admit by University Rating')
plt.xlabel('University Rating')
plt.ylabel('Average Chance of Admit')
plt.show()
```

```python
# 3. Box Plot: TOEFL Score by Research Experience
plt.figure(figsize=(8, 6))
sns.boxplot(x='Research', y='TOEFL Score', data=data)
plt.title('TOEFL Score by Research Experience')
plt.show()
```

```python
# 4. Pie Chart: Distribution of University Ratings
university_rating_counts = data['University Rating'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(university_rating_counts, labels=university_rating_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of University Ratings')
plt.show()
```

```python
# 5. Line Chart: Chance of Admit vs. SOP
plt.figure(figsize=(8, 6))
sns.lineplot(x='SOP', y='Chance of Admit ', data=data)
plt.title('Chance of Admit vs. SOP')
plt.show()
```
